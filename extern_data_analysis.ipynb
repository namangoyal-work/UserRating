```python
from nltk.tokenize import word_tokenize
import nltk
nltk.data.path.append("nltk_data")
from tqdm.notebook import tqdm

tok_set = set()

tok_reviews = []

for idx, review in tqdm(train_df.iterrows()):
    toks = word_tokenize(review[0])
    tok_reviews.append((toks,review[1]))
    for tok in toks:
        tok_set.add(tok)
```

```python
tok_set = set()
for review in tok_reviews:
    for tok in review[0]:
        tok_set.add(tok.lower())
```

```python
import numpy as np

toks = list(tok_set)
rev_toks = {tok:i for i,tok in enumerate(toks)}
tok_class_freq = np.zeros((5,len(toks)))

for review in tqdm(tok_reviews):
    for tok in review[0]:
        tok_class_freq[review[1]-1,rev_toks[tok.lower()]] += 1
```

```python
# zipf law
tok_freq = tok_class_freq.sum(axis=0)
plt.plot(np.sort(tok_freq)[::-1][:400])
```

```python
# better seen via a q-q plot
plt.yscale('log')
plt.xscale('log')
for c in range(5):
    plt.plot(np.sort(tok_class_freq[c,:]/tok_class_freq[c,:].sum())[::-1])
plt.loglog([1, 1e5], [1e-1, 1e-6], '--')
plt.plot([],[],)
```

```python
# need to do this eyeballing + trimming across all classes individually as train data is so biased
# then take union of the stuff
# Also post negation...
tok_freq_lists = [sorted([(f,i) for i,f in enumerate(tok_class_freq[c])])[::-1][10:1000] for c in range(5)]
tok_freqs = set()
for tfl in tok_freq_lists:
    tok_freqs = tok_freqs.union(set([i for f,i in tfl]))
```

```python
len(tok_freqs)
```

```python
# extend this seed set of unigrams to bigrams and trigrams as well
```

```python
def P_w(w):
    return (tok_class_freq[:,w].sum()+5)/(tok_class_freq.sum()+5*len(toks))

def P_w_c(w, c):
    return (tok_class_freq[c,w]+1)/(tok_class_freq[c,:].sum()+len(toks))

def plot_tok_sentiment(tok):
    if type(tok) == str:
        tok_idx = rev_toks[tok]
    else:
        tok_idx = tok
    ll_c = np.array([P_w_c(tok_idx, c) for c in range(5)])
    ll = ll_c/P_w(tok_idx)
    plt.plot(range(1,6), ll, marker='o')
    plt.xticks(range(1,6))
```

```python
plot_tok_sentiment("compliments")
```

```python
# sort based on likelihood ratio
tok_lls = [([P_w_c(i,c)/P_w(i) for c in range(5)],i) for i in tok_freqs]
tok_ll_ratios = [(max(l)/min(l),i) for l,i in tok_lls]
```

```python
top_ratios = sorted(tok_ll_ratios)[::-1]
```

```python
fig, axs = plt.subplots(5,5,figsize=(20,20),dpi=100)

idx = 0
for llr,i in top_ratios[:25]:
    ax = axs[idx//5,idx%5]
    ax.plot(range(1,6), [P_w_c(i,c)/P_w(i) for c in range(5)], marker='.')
    ax.set_title(f'Likelihood for {toks[i]}')
    idx += 1
```
